# **● Create basic visualizations:**


○ Scatter plot to analyze relationships between variables.

```python
import matplotlib.pyplot as plt
# Set figure size (width=5, height=5)
plt.figure(figsize=(5, 3))
plt.scatter(df['SepalLengthCm'],df['SepalWidthCm'])
# Add labels and title
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.title('Scatter Plot of sepal length and width')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Set figure size (width=5, height=5)
plt.figure(figsize=(5, 3))
plt.scatter(df['PetalLengthCm'],df['PetalWidthCm'])
# Add labels and title
plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')
plt.title('Scatter Plot of Petal length and width')
plt.show()
```

Histogram to examine data distribution.

```python
import matplotlib.pyplot as plt
# Set figure size (width=5, height=5)
plt.figure(figsize=(5, 3))
plt.hist(df['SepalLengthCm'])

# Add labels and title
plt.xlabel('SepalLength')
plt.ylabel('Frequency')
plt.title('Histogram of sepal length')
plt.show()
```

○ Box plot to detect outliers and spread of values.

```python
import matplotlib.pyplot as plt
# Set figure size (width=10, height=6)
plt.figure(figsize=(10, 6))
# Create a box plot
plt.boxplot([df['SepalLengthCm'], df['SepalWidthCm'], df['PetalWidthCm'], df['PetalLengthCm']],
            labels=['SepalLengthCm', 'SepalWidthCm', 'PetalWidthCm', 'PetalLengthCm'],
            # Fills the box with color
            patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='green'),
            capprops=dict(color='black'),
            flierprops=dict(marker='x', color='red', markersize=6))

# Add title and show
plt.title('Box plot of Iris Data Set')
plt.grid(True)
plt.show()
```
